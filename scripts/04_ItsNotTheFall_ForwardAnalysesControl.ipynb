```python
from nilearn.plotting import plot_stat_map
from nilearn.image import threshold_img
```

```python
statmap_rmse  = '../neuroscout-9b2o7/fitlins/level-dataset_name-dataset_contrast-rms_stat-z_statmap.nii.gz'

thresholded_map_rmse, threshold_rmse = threshold_stats_img(statmap_rmse, 
                                                                       alpha=0.01, 
                                                                       height_control="fdr"
)
print(f"The FDR = .01 threshold is {threshold_rmse:.3g}")


plot_stat_map( statmap_rmse, 
               threshold  = threshold_rmse,
               #cut_coords = [50, -58, 20],
               title = 'RMSE-Analysis ');
```

```python
statmap_laughter   = '../neuroscout-mz3i4/fitlins/level-dataset_name-dataset_contrast-laughterdummy_stat-t_statmap.nii.gz'

thresholded_map_laughter, threshold_laughter = threshold_stats_img(statmap_laughter, 
                                                                   alpha=0.01, 
                                                                   height_control="fdr"
)
print(f"The FDR = .01 threshold is {threshold_laughter:.3g}")


plot_stat_map( thresholded_map_laughter, 
               threshold  = threshold_laughter,
               cut_coords = [50, -58, 20],
               title = 'Laughter-Analysis - Focus: rTPJ');
```
